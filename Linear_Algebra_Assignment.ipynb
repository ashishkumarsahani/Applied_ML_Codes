```
bwimg = np.zeros([data.shape[0], data.shape[1]], dtype = data.dtype)
for i in range(0, data.shape[0]):
  for j in range(0, data.shape[1]):
    M  = round((0.33 * data[i, j,0]) + (0.33 * data[i, j,1]) + (0.33 * data[i, j,2]));
    bwimg[i,j] = M;
```

```
#Thresholding
Threshold_Val = 150;
threshold_img = np.zeros(bwimg.shape)
for i in range(0, bwimg.shape[0]):
  for j in range(0, bwimg.shape[1]):
    if(bwimg[i,j]<Threshold_Val):
      threshold_img[i,j] = 0;
    else:
      threshold_img[i,j] = 255;

pyplot.imshow(threshold_img, cmap = 'gray')
```

```
#Sobel filtering
filter_Kernel = np.array([[-1, 0, 1], [-2, 0, 2], [-1,0,1]])
print(filter_Kernel)
```

```
filtered_img = np.zeros(bwimg.shape)
for i in range(1, bwimg.shape[0]-1):
  for j in range(1, bwimg.shape[1]-1):
    extracted_array_from_img = bwimg[i-1:i+2, j-1:j+2];
    filtered_img[i,j] = np.sum(extracted_array_from_img*filter_Kernel)

```

```
pyplot.imshow(filtered_img, cmap = 'gray')
```
